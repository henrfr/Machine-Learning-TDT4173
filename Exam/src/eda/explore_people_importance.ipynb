```python
import pandas as pd
from dython.nominal import associations
```

```python
people = pd.read_csv("../../own_data/grunnkrets_norway_large.csv")
stores = pd.read_csv("../../data/stores_train.csv")
```

```python
df = stores.merge(people, how="left", on=["grunnkrets_id", "year"])
```

```python
df = df.drop(columns=["store_id", "lat", "lon", "geometry", "plaace_hierarchy_id", "year", "store_name", "address"])
```

```python
df = df.drop(columns=["grunnkrets_age_0-14", "grunnkrets_age_15-64", "grunnkrets_age_65-90", "district_age_0-14", "district_age_15-64", "district_age_65-90"
, "municipality_age_0-14", "municipality_age_15-64", "municipality_age_65-90"])
df = df.drop(columns=["grunnkrets_pop_number", "district_pop_number", "municipality_pop_number"])
```
